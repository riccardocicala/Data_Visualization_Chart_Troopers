```python
nodes = []

for i, year in enumerate(df4_final_energy_consumption_by_product_with_total['year'].unique()):
    n_nodes = len(nodes)

    df4_final_energy_consumption_by_sector_with_total_by_sector_curr_year = df4_final_energy_consumption_by_sector_with_total_by_sector[df4_final_energy_consumption_by_sector_with_total_by_sector['year'] == year].reset_index(drop=True)
    df5_final_energy_consumption_by_sector_with_total_by_country_curr_year = df5_final_energy_consumption_by_sector_with_total_by_country[df5_final_energy_consumption_by_sector_with_total_by_country['year'] == year].reset_index(drop=True)
    df4_final_energy_consumption_by_product_with_total_curr_year = df4_final_energy_consumption_by_product_with_total[df4_final_energy_consumption_by_product_with_total['year'] == year].reset_index(drop=True)

    curr_n_sectors = df4_final_energy_consumption_by_sector_with_total_by_sector_curr_year['sector_type'].nunique()
    curr_n_countries = df5_final_energy_consumption_by_sector_with_total_by_country_curr_year['country'].nunique()
    curr_n_products = df4_final_energy_consumption_by_product_with_total_curr_year['fuel_type'].nunique()

    for idx, row in df4_final_energy_consumption_by_sector_with_total_by_sector_curr_year.iterrows():
        nodes.append(pd.DataFrame({"node" : [row["sector_type"]], "value" : [row["total_sector_consumption_value"]], "year" : [row["year"]], "indirect_links" : [[n for n in range(curr_n_sectors + curr_n_countries, curr_n_sectors + curr_n_countries + curr_n_products)]], "idx" : idx}))

    for idx, row in df5_final_energy_consumption_by_sector_with_total_by_country_curr_year.iterrows():
        nodes.append(pd.DataFrame({"node" : [row["country"]], "value" : [row["total_country_consumption_value"]], "year" : [row["year"]], "indirect_links" : [[]], "idx" : idx + curr_n_sectors}))

    for idx, row in df4_final_energy_consumption_by_product_with_total_curr_year.iterrows():
        nodes.append(pd.DataFrame({"node" : [row["fuel_type"]], "value" : [row["total_product_consumption_value"]], "year" : [row["year"]], "indirect_links" : [[n for n in range(0 , curr_n_sectors)]], "idx" : idx + curr_n_sectors + curr_n_countries}))

nodes = pd.concat(nodes + []).reset_index(drop=True)

nodes['value'] = (nodes['value'] / 1_000).round(2)

nodes = pd.merge(nodes, df_nodes_colors, how="inner")

nodes.to_csv('prepared_datasets/plot1_nodes.csv')
nodes
```

```python
df5_final_energy_consumption_by_product = pd.merge(df5_final_energy_consumption_by_product, df5_final_energy_consumption_by_sector_with_total_by_country, how = "inner", on = ["country", "year"])
df5_final_energy_consumption_by_product
```

```python
df_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = df2_final_energy_consumption_in_transport_by_type_of_fuel[["year", "country", "consumption_value"]].groupby(["country", "year"]).sum().rename(columns={"consumption_value" : "total_country_consumption_value"})
df2_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = df_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.sort_values(by=['year', 'total_country_consumption_value'], ascending=[True, False]).reset_index()
df2_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year
```

```python
df3_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = (df2_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.groupby('year').apply(lambda x: x.nlargest(top_n, 'total_country_consumption_value'))).reset_index(drop=True)
df3_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year
```

```python
df4_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = pd.merge(df3_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year, df2_final_energy_consumption_in_transport_by_type_of_fuel, how = "left", on = ["country", "year"])
df4_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year
```

```python
df5_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = pd.merge(df4_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year, df3_final_energy_consumption_in_transport_by_type_of_fuel_total_per_year, how = "left", on = ["fuel_type", "year"])
df5_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year[:30]
```

```python
df6_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = df5_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.copy()

df6_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year['fuel_type'] = df6_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.apply(
    lambda row: 'Other' if pd.isnull(row['total_type_of_fuel_consumption_value']) else row['fuel_type'], 
    axis=1
)

df6_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year
```

```python
# Define the sorting function for 'fuel_type'
def custom_sort_fuel_type(fuel_type):
    return 1 if fuel_type == "Other" else 0  # Assign a higher rank to "Other"

df7_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = df6_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.groupby(
    ["country", "year", "total_country_consumption_value", "fuel_type", "total_type_of_fuel_consumption_value"]).sum().reset_index()

# Add a helper column to sort 'fuel_type'
df7_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year["fuel_type_rank"] = df7_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year["fuel_type"].map(custom_sort_fuel_type)

# Perform the sort
df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = df7_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.sort_values(
    by=["year", "total_country_consumption_value", "fuel_type_rank", "consumption_value"],
    ascending=[True, False, True, False]  # Sort 'fuel_type_rank' ascending to place 'Other' last
)

# Drop the helper column as it's no longer needed
df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.drop(columns=["fuel_type_rank"]).reset_index(drop=True)



df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.head(30)
```

```python
df9_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year
df9_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year['Rank'] = df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.groupby(['country', 'year'])['consumption_value'].rank(method='first', ascending=False)
df9_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.head(30)
```

```python
df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year['consumption_value'] = (df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year['consumption_value'] / 1_000).round(2)
df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year
```

```python
df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.to_csv('prepared_datasets/plot3.csv')
```

```python
selected_columns_waste = ["wst_oper", "TIME_PERIOD", "OBS_VALUE"]
df2_total_waste_management_by_type = df_total_waste_management_by_type[selected_columns_waste].rename(columns = {"wst_oper" : "waste_operation", "TIME_PERIOD" : "year", "OBS_VALUE" : "waste_disposed_in_tonnes"})
df2_total_waste_management_by_type
```

```python
df2_total_waste_management_by_type = df2_total_waste_management_by_type[df2_total_waste_management_by_type["waste_operation"] != "Waste treatment"]
```

```python
# Add a column for operation type (Disposal/Recovery)
df2_total_waste_management_by_type['operation_type'] = df2_total_waste_management_by_type['waste_operation'].str.split(' - ').str[0]

# Group by year and operation type, calculate the mean
df3_total_waste_management_by_type = (
    df2_total_waste_management_by_type.groupby(['year', 'operation_type'])['waste_disposed_in_tonnes']
    .mean()
    .reset_index())

# Append the calculated averages to the original DataFrame
df3_total_waste_management_by_type['waste_operation'] = df3_total_waste_management_by_type['operation_type'] + " - average"
df4_total_waste_management_by_type = pd.concat([df2_total_waste_management_by_type, df3_total_waste_management_by_type], ignore_index=True)

# Sort by year and reset index

df4_total_waste_management_by_type
```

```python
df4_total_waste_management_by_type['waste_disposed_in_million_tonnes'] = (df4_total_waste_management_by_type['waste_disposed_in_tonnes'] / 1_000_000).round(2)
df4_total_waste_management_by_type
```
