```python
import pandas as pd
import numpy as np
```

## Plot 1

```python
df_final_energy_consumption_by_product = pd.read_csv("datasets/final_energy_consumption_by_product.csv")
df_final_energy_consumption_by_sector = pd.read_csv("datasets/final_energy_consumption_by_sector.csv")
```

```python
selected_columns_product = ["siec", "geo", "TIME_PERIOD", "OBS_VALUE"]
selected_columns_sector = ["nrg_bal", "geo", "TIME_PERIOD", "OBS_VALUE"]
df2_final_energy_consumption_by_product = df_final_energy_consumption_by_product[selected_columns_product]
df2_final_energy_consumption_by_sector = df_final_energy_consumption_by_sector[selected_columns_sector]
```

```python
df3_final_energy_consumption_by_product = df2_final_energy_consumption_by_product.rename(columns={"siec":"fuel_type", "geo":"country", "TIME_PERIOD":"year", "OBS_VALUE":"consumption_value"}).reset_index(drop=True)
df3_final_energy_consumption_by_sector = df2_final_energy_consumption_by_sector.rename(columns={"nrg_bal":"sector_type", "geo":"country", "TIME_PERIOD":"year", "OBS_VALUE":"consumption_value"})
```

```python
df3_final_energy_consumption_by_product
```

```python
df3_final_energy_consumption_by_sector
```

```python
df4_final_energy_consumption_by_product = df3_final_energy_consumption_by_product[df3_final_energy_consumption_by_product["fuel_type"] != "Total"].reset_index(drop=True)
df4_final_energy_consumption_by_product
```

```python
df4_final_energy_consumption_by_sector = df3_final_energy_consumption_by_sector[df3_final_energy_consumption_by_sector["sector_type"] != "Final consumption - energy use"].reset_index(drop=True)
df4_final_energy_consumption_by_sector
```

```python
df4_final_energy_consumption_by_product_with_total = df4_final_energy_consumption_by_product[["fuel_type", "year", "consumption_value"]].groupby(["fuel_type", "year"]).sum().reset_index().rename(columns={"consumption_value" : "total_product_consumption_value"})
df4_final_energy_consumption_by_product_with_total
```

```python
df4_final_energy_consumption_by_sector_with_total_by_sector = df4_final_energy_consumption_by_sector[["sector_type", "year", "consumption_value"]].groupby(["sector_type", "year"]).sum().reset_index().rename(columns={"consumption_value" : "total_sector_consumption_value"})
df4_final_energy_consumption_by_sector_with_total_by_sector
```

```python
df4_final_energy_consumption_by_sector_with_total_by_country = df4_final_energy_consumption_by_sector[["country", "year", "consumption_value"]].groupby(["country", "year"]).sum().reset_index().rename(columns={"consumption_value" : "total_country_consumption_value"})
df4_final_energy_consumption_by_sector_with_total_by_country
```

```python
df5_final_energy_consumption_by_product = pd.merge(df4_final_energy_consumption_by_product, df4_final_energy_consumption_by_product_with_total, how = "left", on = ["fuel_type", "year"])
df5_final_energy_consumption_by_product
```

```python
df5_final_energy_consumption_by_sector = pd.merge(df4_final_energy_consumption_by_sector, df4_final_energy_consumption_by_sector_with_total_by_sector, how = "left", on = ["sector_type", "year"])
df5_final_energy_consumption_by_sector
```

```python
nodes = []
for idx, row in df4_final_energy_consumption_by_sector_with_total_by_sector.iterrows():
    nodes.append(pd.DataFrame({"node" : [row["sector_type"]], "value" : [row["total_sector_consumption_value"]], "year" : [row["year"]]}))

for idx, row in df4_final_energy_consumption_by_sector_with_total_by_country.iterrows():
    nodes.append(pd.DataFrame({"node" : [row["country"]], "value" : [row["total_country_consumption_value"]], "year" : [row["year"]]}))

for idx, row in df4_final_energy_consumption_by_product_with_total.iterrows():
    nodes.append(pd.DataFrame({"node" : [row["fuel_type"]], "value" : [row["total_product_consumption_value"]], "year" : [row["year"]]}))

nodes = pd.concat(nodes + []).reset_index(drop=True)
nodes.to_csv('prepared_datasets/plot1_nodes.csv')
nodes
```

```python
links = []
for idx, row in df5_final_energy_consumption_by_sector.iterrows():
    links.append(pd.DataFrame({"source" : [nodes.index[(nodes["node"] == row["sector_type"]) & (nodes["year"] == row["year"])].item()], "target" : [nodes.index[(nodes["node"] == row["country"]) & (nodes["year"] == row["year"])].item()], "value" :  row["consumption_value"]}))

for idx, row in df5_final_energy_consumption_by_product.iterrows():
    links.append(pd.DataFrame({"source" : [nodes.index[(nodes["node"] == row["country"]) & (nodes["year"] == row["year"])].item()], "target" : [nodes.index[(nodes["node"] == row["fuel_type"]) & (nodes["year"] == row["year"])].item()], "value" :  row["consumption_value"]}))

links = pd.concat(links + []).reset_index(drop=True)
links.to_csv('prepared_datasets/plot1_links.csv')
links
```

## Plot 2

```python
df_final_energy_consumption_in_transport_by_type_of_fuel = pd.read_csv("datasets/final_energy_consumption_in_transport_by_type_of_fuel.csv")
df_final_energy_consumption_in_transport_by_type_of_fuel
```

```python
selected_columns_transport_by_type_of_fuel = ["siec", "geo", "TIME_PERIOD", "OBS_VALUE"]
df2_final_energy_consumption_in_transport_by_type_of_fuel = df_final_energy_consumption_in_transport_by_type_of_fuel[selected_columns_transport_by_type_of_fuel].rename(columns = {"siec" : "fuel_type", "geo" : "country", "TIME_PERIOD" : "year", "OBS_VALUE" : "consumption_value"})
df2_final_energy_consumption_in_transport_by_type_of_fuel = df2_final_energy_consumption_in_transport_by_type_of_fuel[df2_final_energy_consumption_in_transport_by_type_of_fuel["fuel_type"] != "Total"]
df2_final_energy_consumption_in_transport_by_type_of_fuel
```

```python
df_final_energy_consumption_in_transport_by_type_of_fuel_total_per_year = df2_final_energy_consumption_in_transport_by_type_of_fuel[["fuel_type", "year", "consumption_value"]].groupby(["fuel_type", "year"]).sum().reset_index().rename(columns={"consumption_value" : "total_type_of_fuel_consumption_value"})
df_final_energy_consumption_in_transport_by_type_of_fuel_total_per_year
```

```python
df2_final_energy_consumption_in_transport_by_type_of_fuel_total_per_year = df_final_energy_consumption_in_transport_by_type_of_fuel_total_per_year.sort_values(["year", "total_type_of_fuel_consumption_value"], ascending=[True, False]).reset_index(drop=True)
df2_final_energy_consumption_in_transport_by_type_of_fuel_total_per_year
```

```python
top_n = 5
df3_final_energy_consumption_in_transport_by_type_of_fuel_total_per_year = (df2_final_energy_consumption_in_transport_by_type_of_fuel_total_per_year.groupby('year').apply(lambda x: x.nlargest(top_n, 'total_type_of_fuel_consumption_value'))).reset_index(drop=True)
df3_final_energy_consumption_in_transport_by_type_of_fuel_total_per_year
```

```python
df3_final_energy_consumption_in_transport_by_type_of_fuel_total_per_year['total_type_of_fuel_consumption_value'] = (df3_final_energy_consumption_in_transport_by_type_of_fuel_total_per_year['total_type_of_fuel_consumption_value'] / 1_000).round(2)
df3_final_energy_consumption_in_transport_by_type_of_fuel_total_per_year
```

```python
df3_final_energy_consumption_in_transport_by_type_of_fuel_total_per_year.to_csv('prepared_datasets/plot2.csv')
```

## Plot 3

```python
df_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = df2_final_energy_consumption_in_transport_by_type_of_fuel[["year", "country", "consumption_value"]].groupby(["country", "year"]).sum().rename(columns={"consumption_value" : "total_country_consumption_value"})
df2_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = df_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.sort_values(by=['year', 'total_country_consumption_value'], ascending=[True, False]).reset_index()
df2_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year
```

```python
df3_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = (df2_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.groupby('year').apply(lambda x: x.nlargest(top_n, 'total_country_consumption_value'))).reset_index(drop=True)
df3_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year
```

```python
df4_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = pd.merge(df3_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year, df2_final_energy_consumption_in_transport_by_type_of_fuel, how = "left", on = ["country", "year"])
df4_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year
```

```python
df5_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = pd.merge(df4_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year, df3_final_energy_consumption_in_transport_by_type_of_fuel_total_per_year, how = "left", on = ["fuel_type", "year"])
df5_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year[:30]
```

```python
df6_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = df5_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.copy()

df6_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year['fuel_type'] = df6_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.apply(
    lambda row: 'Other' if pd.isnull(row['total_type_of_fuel_consumption_value']) else row['fuel_type'], 
    axis=1
)

df6_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year
```

```python
# Define the sorting function for 'fuel_type'
def custom_sort_fuel_type(fuel_type):
    return 1 if fuel_type == "Other" else 0  # Assign a higher rank to "Other"

df7_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = df6_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.groupby(
    ["country", "year", "total_country_consumption_value", "fuel_type", "total_type_of_fuel_consumption_value"]).sum().reset_index()

# Add a helper column to sort 'fuel_type'
df7_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year["fuel_type_rank"] = df7_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year["fuel_type"].map(custom_sort_fuel_type)

# Perform the sort
df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = df7_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.sort_values(
    by=["year", "total_country_consumption_value", "fuel_type_rank", "consumption_value"],
    ascending=[True, False, True, False]  # Sort 'fuel_type_rank' ascending to place 'Other' last
)

# Drop the helper column as it's no longer needed
df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.drop(columns=["fuel_type_rank"]).reset_index(drop=True)



df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.head(30)
```

```python
df9_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year = df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year
df9_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year['Rank'] = df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.groupby(['country', 'year'])['consumption_value'].rank(method='first', ascending=False)
df9_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.head(30)
```

```python
df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year['consumption_value'] = (df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year['consumption_value'] / 1_000).round(2)
df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year
```

```python
df8_final_energy_consumption_in_transport_by_type_of_fuel_top_countries_per_year.to_csv('prepared_datasets/plot3.csv')
```
